```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x) * np.exp(-0.1 * x)

# Generate x values
x = np.linspace(0, 20, 400)
y = f(x)


plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$y = \sin(x) e^{-0.1x}$', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Non-Polynomial Function')
plt.legend()
plt.grid()
plt.show()
```
